```python
# your code here

# Importing the required libraries

import pandas as pd        # For data manipulation and analysis
import numpy as np         # For numerical computations
import matplotlib.pyplot as plt  # For data visualization
import seaborn as sns      # For statistical data visualization
import random              # For generating random numbers

# Optional: Setting up the aesthetics for seaborn
sns.set(style='whitegrid')

# Optional: To display plots inline if using Jupyter notebooks
%matplotlib inline
```

```python
# Set random seed for reproducibility
np.random.seed(50)

# Number of samples
num_samples = 1500

# Generate random data
data = {
    'tweet_id': np.arange(1, num_samples + 1),
    'user_id': np.random.randint(1, 300, num_samples),
    'category': np.random.choice(['News', 'Sports', 'Entertainment', 'Politics', 'Technology', 'Food'], num_samples),
    'likes': np.random.randint(0, 1000, num_samples),
    'retweets': np.random.randint(0, 500, num_samples),
    'text': ['A sample tweet' for _ in range(num_samples)]
}

# Create DataFrame
tweets_df = pd.DataFrame(data)

# Display the first few rows of the dataframe
tweets_df.head()
```

```python
# Display basic statistics
print(tweets_df.describe())

# Display distribution of likes
plt.figure(figsize=(10, 6))
sns.distplot(tweets_df['likes'], bins=30, kde=True)
plt.title('Distribution of Likes')
plt.xlabel('Number of Likes')
plt.ylabel('Frequency')
plt.show()

# Display distribution of tweets by category
plt.figure(figsize=(10, 6))
sns.countplot(x='category', data=tweets_df)
plt.title('Number of Tweets by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
# Average likes per category
avg_likes_per_category = tweets_df.groupby('category')['likes'].mean().reset_index()

# Average retweets per category
avg_retweets_per_category = tweets_df.groupby('category')['retweets'].mean().reset_index()

# Plotting average likes per category
plt.figure(figsize=(10, 6))
sns.barplot(x='category', y='likes', data=avg_likes_per_category)
plt.title('Average Likes per Category')
plt.xlabel('Category')
plt.ylabel('Average Likes')
plt.show()

# Plotting average retweets per category
plt.figure(figsize=(10, 6))
sns.barplot(x='category', y='retweets', data=avg_retweets_per_category)
plt.title('Average Retweets per Category')
plt.xlabel('Category')
plt.ylabel('Average Retweets')
plt.show()
```
